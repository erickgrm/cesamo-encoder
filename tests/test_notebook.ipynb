```python
import numpy as np
import pandas as pd
import sys
sys.path.append('../')
from cesamo_encoder import CESAMOEncoder
#import warnings
#warnings.filterwarnings("ignore")
```

```python
# Example dataset
X = pd.DataFrame([['x',0.4,0.7,'a'],['y',0.1,0.9,'b'], ['x',0.2,0.83,'b'], ['x',0.14,0.78,'b']])
y = pd.Series([1, 0, 1,0])

# Load encoder, fit_transform X
enc =  CESAMOEncoder(estimator_name='CESAMORegression', plot=True)
X_enc = enc.fit_transform(X,y)

# Print datasets
print('Original dataset\n', pd.concat((X,y),axis=1).values)
print('Encoded dataset\n', pd.concat((X_enc,y),axis=1).values)
```
